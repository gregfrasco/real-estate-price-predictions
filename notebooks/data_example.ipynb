```python
def encodeColumn(df, column, providedList=None):
    if(providedList == None):
        providedList = list(df[column].unique())
    notInListValue = len(providedList)
    return [providedList.index(x) if x in providedList else notInListValue for x in df[column]]
```

```python
df['PROPTYPE_CAT'] = encodeColumn(df, 'PROPTYPE', ['SF', 'CC', 'MF'])
df[['PROPTYPE','PROPTYPE_CAT']].head(20)
```

```python
df['STYLE_CAT'] = encodeColumn(df, 'STYLE')
df[['STYLE','STYLE_CAT']].head(20)
```

```python
df['ZIP_CAT'] = encodeColumn(df, 'ZIP')
df[['ZIP','ZIP_CAT']].head(20)
```

```python
def convertDate(date, dateFormat='%m/%d/%Y'):
    if(isinstance(date, str)):
        return int(datetime.strptime(date, dateFormat).strftime("%Y%m%d"))
    return date
```

```python
df['LISTDATE_CAT'] = [convertDate(x) for x in df['LISTDATE']]
```

```python
df['SOLDDATE_CAT'] = [convertDate(x) for x in df['SOLDDATE']]
```
