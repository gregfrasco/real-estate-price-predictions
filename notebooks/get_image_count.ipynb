```python
#imports
import pandas as pd
import numpy as np
```

## Get image names from image meta data
image file meta data contained in image_name text file. Filter out just the MLS number from the image names.

```python
df = pd.read_csv('../data/image_names.txt',header=None)
df.rename(columns={0:'image_info'},inplace=True)
imgs = df[df['image_info'].str.contains('.jpg')]
imgs = imgs["image_info"].str.split('_').str[0].reset_index()
imgs.rename(columns={'image_info':'MLS_num'},inplace=True)
imgs
```

## Count images
Count images with the same MLS number and add as a separate column

```python
image_count = imgs['MLS_num'].value_counts().reset_index()
image_count = image_count.rename(columns={'MLS_num':'count','index':'MLS_num'})
image_count
```

```python
image_count.to_csv('image_count.csv')
```
